```python
names = ["anne hathaway", "robert de niro", "nancy meyers"]

# 이름에서 "hathaway", "de", "niro"를 제거
pattern = r'\bhathaway\b|\bde\b|\bniro\b|\bmeyers\b|\bmeyer\b|\bdeniro\b|\bwhittaker\b|\bostin\b|\bdeniros\b|\bhathaways\b|\bniros\b|\bdiniro\b'


def align_name(text):

    # "deniros"를 "robert"로 대체
    text = re.sub(r'\bdeniros\b', 'robert', text)
    # "hathaways"를 "anne"로 대체
    text = re.sub(r'\bhathaways\b', 'anne', text)
    # "de niro"를 "robert"로 대체
    text = re.sub(r'\bde niro\b', 'robert', text)
    # "robert de niro"를 "robert"로 대체
    text = re.sub(r'\brobert de niro\b', 'robert', text)
    text = re.sub(r'\bniros\b', 'robert', text)
    text = re.sub(r'\diniro\b', 'robert', text)
    #나머지는 다 삭제
    text = re.sub(pattern, '', text)
    return text
```

```python
def clean_text(text):
    if isinstance(text, datetime):
        return ''
    return re.sub('[^.,?!\s\w]', '', text)
```

```python
import re

def remove_special_chars(text):
    if isinstance(text, str):
        pattern = r'[^a-zA-Z0-9\s]'
        cleaned_text = re.sub(pattern, '', text)
        return cleaned_text
    else:
        return text

america_review_df['cleaned_sentence'] = america_review_df['splited_sentence'].apply(remove_special_chars)
```

```python
#불용어 삭제하기

result = []
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

clean_review_series = pd.Series(clean_review)
clean_review_series = clean_review_series.apply(remove_stopwords)
```

```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence
from nltk import bigrams

# clean_review를 시리즈로 변환
clean_review_series = pd.Series(clean_review)

# 텍스트를 토큰화하고 bigram을 생성하는 함수
def tokenize_and_create_bigrams(text):
    # 토큰화
    tokens = text_to_word_sequence(text)
    # bigram 생성
    bigram_list = list(bigrams(tokens))
    return bigram_list

# clean_review_series에 적용
bigrams_series = clean_review_series.apply(tokenize_and_create_bigrams)

```
